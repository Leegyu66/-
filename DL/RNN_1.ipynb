## 순차 데이터
- 텍스트나 시계열 데이터(일정한 시간 간격으로 기록된 데이터)와 같이 순서에 의미가 있는 데이터를 말한다.
- ex) I am a boy 와 같이 순서가 중요한 데이터 
- 이러한 유형의 데이터를 다룰 때는 이전에 입력한 데이터를 기억하는 기능이 필요하다
- ex) "별로지만 추천해요" 라는 데이터가 들어왔을 때 "별로지만"을 기억해야지 이 댓글을 무조건 긍정적이라고 판단하지 않을 것임

- 하지만 완전 연결 신경망이나 합성곱 신경망은 이런 기억 장치가 없다
- 하나의 샘플을 사용하여 정방향 계산을 하고 나면 그 샘플은 버려지고 다음 샘플을 처리할 때  재사용하지 않는다
